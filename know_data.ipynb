### construct distance dict of from_AQStations_to_WeatherStation

```python
# 通过air quality站点的经纬度，定位四个grid所形成的区域
# 获取air quality站点经纬度信息
aq_geo_info = pd.read_csv("./data/MSBD5002PROJECT_data/Beijing_air_quality_station.csv",index_col=None)
# 获取grid weather站点经纬度信息
gw_geo_info = pd.read_csv("./data/MSBD5002PROJECT_data/Beijing_grid_weather_station.csv",header=None,names=["gw_name","gw_latitude","gw_longitude"])
# 获取observed weather站点经纬度信息
ow_geo_info = pd.read_csv("./data/MSBD5002PROJECT_data/Beijing_observed_weather_station.csv",header=None,names=["ow_name","ow_longitude","ow_latitude"])
```

```python
def find_nearest_gw(df):
    # original long & lat
    aq_longitude = df["longitude"]
    aq_latitude = df["latitude"]
    
    # rounded long & lat
    # caution! np.round 会四舍五入
    round_aq_longtitude = np.round(aq_longitude,decimals=1)
    round_aq_latitude = np.round(aq_latitude,decimals=1)
    
    # use rounded data to find nearest grid weather station
    middle_result = gw_geo_info[gw_geo_info["gw_longitude"] == round_aq_longtitude]
    grid_st = middle_result[middle_result["gw_latitude"] == round_aq_latitude]
    
    # add grid_weather information into new dataframe
    df["nearest_gw"] = grid_st["gw_name"].iloc[0]
    df["gw_latitude"] = grid_st["gw_latitude"].iloc[0]
    df["gw_longitude"] = grid_st["gw_longitude"].iloc[0]
    
    # calculate distance between nearest grid weather station and this air quality station
    df["dist_gw"] = np.power((round_aq_longtitude - aq_longitude),2) + \
                    np.power((round_aq_latitude - aq_latitude),2)
    
    return df
```

```python
# for one air quality station
# generate all distance 
def helper(ow_df,lo,la):
    ow_longitude = ow_df["ow_longitude"]
    ow_latitude = ow_df["ow_latitude"]
    
    # calculate distance between  weather station and this air quality station
    ow_df["dist_ow"] = np.power((lo-ow_longitude),2) + \
                       np.power((la-ow_latitude),2)
    return ow_df

def find_nearest_ow(df):
    # original long & lat
    aq_longitude = df["longitude"]
    aq_latitude = df["latitude"]
    
    # generate a dataframe: of 1 air quality station, distances from all observed weather stations
    tmp = ow_geo_info.apply(lambda row: helper(row,aq_longitude,aq_latitude),axis=1)
    # pick one observed weather station which has the shortest distance to "the" air_quality station 
    nearest_ow_df = tmp[tmp["dist_ow"] == tmp["dist_ow"].min()]
    
    # add observed_weather information into new dataframe
    df["nearest_ow"] = nearest_ow_df["ow_name"].iloc[0]
    df["ow_latitude"] = nearest_ow_df["ow_latitude"].iloc[0]
    df["ow_longitude"] = nearest_ow_df["ow_longitude"].iloc[0]
    
    df["dist_ow"] = nearest_ow_df["dist_ow"].iloc[0]
    
    return df
```

```python
# 遍历air quality站点从而获取最近的气象站名，用于并表
geo_info = aq_geo_info.apply(find_nearest_gw,axis=1)
geo_info_new = geo_info.apply(find_nearest_ow,axis=1)
```
